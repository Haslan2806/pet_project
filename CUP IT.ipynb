```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

color_pal = sns.color_palette()
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv("CL_Cup IT 2024_Сбер_dataset.csv")
df['ctl_loading_date'] = pd.to_datetime(df['ctl_loading_date'])
```

```python
df.head()
```

```python
grouped = df.groupby('ctl_loading')['benchmark_id'].apply(list).reset_index()
```

```python
grouped
```

```python
#df['ctl_loading_count'] = df.groupby('ctl_loading')['ctl_loading'].transform('count')
```

```python
df.head()
```

```python
date = df['ctl_loading_date']
df['hour'] = date.dt.hour
df['dayofweek'] = date.dt.dayofweek
df['quarter'] = date.dt.quarter
df['month'] = date.dt.month
df['year'] = date.dt.year
df['dayofyear'] = date.dt.dayofyear
df['dayofmonth'] = date.dt.day
df['weekofyear'] = date.dt.isocalendar().week
```

```python
df.head()
```

```python
X = df.drop(columns = ['benchmark_id', 'benchmark_name', 'date_of', 'quote'])
X = X.drop_duplicates()
```

```python
X = pd.merge(X, grouped, on = 'ctl_loading')
```

```python
X.head()
```

```python
X['num_bench'] = X['benchmark_id'].apply(len)
X['num_bench_unique'] = X['benchmark_id'].apply(set).apply(len)
```

```python
def count_benchmark_ids(benchmark_ids):
    all_ids = [item for sublist in benchmark_ids for item in sublist]
    return pd.Series(all_ids).value_counts()

benchmark_counts = count_benchmark_ids(X["benchmark_id"])
for benchmark_id in benchmark_counts.index:
    X[benchmark_id] = 0

for i, row in X.iterrows():
    for benchmark_id in row["benchmark_id"]:
        X.at[i, benchmark_id] += 1
X.drop(columns=['benchmark_id'], inplace = True)
```

```python
X.head(10)
```

```python
X.columns
```

```python
def is_weekend(x):
    return int(x > 4)
X['is_weekend'] = X['dayofweek'].apply(is_weekend)
```

```python
X.shape
```

```python
Q1 = X['ctl_loading_count'].quantile(0.25)
Q3 = X['ctl_loading_count'].quantile(0.75)

IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print([lower, upper]) 
```

```python
def get_target(x):
    return int(lower <= x <= upper)
X['target'] = X['ctl_loading_count'].apply(get_target)
```

```python
features = ['dayofweek', 'quarter',
       'month', 'year', 'dayofyear', 'dayofmonth', 'weekofyear', 'is_weekend']
target = 'ctl_loading_count'
```

```python
X.info()
```

## Train/Test Split

```python
train = X.loc[X.ctl_loading_date < '2023-04-15']
test = X.loc[X.ctl_loading_date >= '2023-04-15']
```

```python
train = train[train['target'] == 1]
```

```python
n_train = train.drop(columns='ctl_loading_date')
n_test = test.drop(columns='ctl_loading_date')
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mtf = StandardScaler()
mtf.fit_transform(n_train)
mtf.fit_transform(n_test)
```

```python
X_train = n_train[features]
y_train = n_train[target]

X_test = n_test[features]
y_test = n_test[target]
ans = n_test['target']
```

## KNN Regressor

```python
model = KNeighborsRegressor(n_neighbors=15)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mean_absolute_error(y_pred, y_test)
```

## Linear Reg

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mean_absolute_error(y_pred, y_test)
```

## Lasso

```python
from sklearn.linear_model import Lasso
model = Lasso(alpha = 10)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mean_absolute_error(y_pred, y_test)
```

## Ridge

```python
from sklearn.linear_model import Ridge
model = Ridge()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mean_absolute_error(y_pred, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(criterion='absolute_error',
                              splitter='best',
                              max_depth=2)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mean_absolute_error(y_pred, y_test)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(criterion='absolute_error',
                              n_estimators=10,
                              max_depth = 2)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mean_absolute_error(y_pred, y_test)
```

```python
from xgboost import XGBRegressor
model = XGBRegressor(n_estimators = 1, eval_metric = 'mae', max_depth = 2, alpha = 10)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mean_absolute_error(y_pred, y_test)
```

```python
def get_accuracy(y_pred, y_test, eps):
    return np.sum((np.abs(y_pred- y_test) < eps) == ans)
```

```python
history = []
for i in range(50, 10, -1):
    history.append(get_accuracy(y_pred, y_test, i)/ y_test.shape[0])
    print(i, history[-1])
data = pd.DataFrame({
    'eps': range(50, 10, -1),
    'accuracy': history
})
sns.lineplot(data = data, x = 'eps', y = 'accuracy')
```

## Voting Regressor

```python
from sklearn.ensemble import VotingRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor

est1 = Lasso(alpha = 10)
est2 = SVR()
est3 = DecisionTreeRegressor(criterion='absolute_error',
                              splitter='best',
                              max_depth=2)
model = VotingRegressor(
    estimators=[('1', est1), 
                ('2',est2), 
                ('3', est3)]
)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mean_absolute_error(y_pred, y_test)
```

```python
from sklearn.ensemble import StackingRegressor

est1 = Lasso(alpha = 10)
est2 = SVR()
est3 = DecisionTreeRegressor(criterion='absolute_error',
                              splitter='best',
                              max_depth=2)
fin = LinearRegression()

model = StackingRegressor(
    estimators=[('lasso', est1),
                ('svr', est2),
                ('dt',est3)],
    final_estimator=fin
)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mean_absolute_error(y_pred, y_test)
```
